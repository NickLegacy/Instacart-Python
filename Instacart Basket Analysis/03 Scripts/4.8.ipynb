# Step 6

## Create a spending flag for each user based on the average price across all their orders

```python
# Creating a Spending flag using loc. New column "average_price". 
orders_products_merged_47['average_price'] = orders_products_merged_47.groupby(['user_id'])['prices'].transform(np.max)
```

```python
orders_products_merged_47.head()
```

```python
# Spending Flag user is lower than 10, then flag them as a “Low spender.”
orders_products_merged_47.loc[orders_products_merged_47['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Spending Flag user is higher than or equal to 10, then flag them as a “High spender.”
orders_products_merged_47.loc[orders_products_merged_47['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

# Step 7

## Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Creating a Frequency flag using loc. New column "median_prior_orders". 
orders_products_merged_47['median_prior_orders'] = orders_products_merged_47.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
# Median in days_since_prior_order column is high than 20, then flag them as a “Non-frequent customer.”
orders_products_merged_47.loc[orders_products_merged_47['median_prior_orders'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Median in days_since_prior_order column is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
orders_products_merged_47.loc[(orders_products_merged_47['median_prior_orders'] > 10) & (orders_products_merged_47['median_prior_orders'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Median in days_since_prior_order column is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
orders_products_merged_47.loc[orders_products_merged_47['median_prior_orders'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
